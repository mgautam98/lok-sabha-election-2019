```python
indicesToDrop = []
for ix in range(len(data)):
    if(data.iloc[ix]['Cases']==0):
        indicesToDrop.append(ix)
```

```python
data.drop(data.index[indicesToDrop], inplace=True)
```

```python
data = data.reset_index()
```

```python
sums = data.Cases.groupby(data.Party).sum()
```
